```python
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid", palette="bright")
```

### Load Saved Results

```python
df_deepGo_CC=pd.read_csv('data/_Outputs/Final_DeepGO_CC.csv')
df_deepGo_BP=pd.read_csv('data/_Outputs/Final_DeepGO_BP.csv')
df_deepGo_MF=pd.read_csv('data/_Outputs/Final_DeepGO_MF.csv')

df_TripletProt_CC=pd.read_csv('data/_Outputs/Final_TripletProt_CC.csv')
df_TripletProt_BP=pd.read_csv('data/_Outputs/Final_TripletProt_BP.csv')
df_TripletProt_MF=pd.read_csv('data/_Outputs/Final_TripletProt_MF.csv')

df_unirep_CC=pd.read_csv('data/_Outputs/Final_Unirep_CC.csv')
df_unirep_BP=pd.read_csv('data/_Outputs/Final_Unirep_BP.csv')
df_unirep_MF=pd.read_csv('data/_Outputs/Final_Unirep_MF.csv')

df_kmers_CC=pd.read_csv('data/_Outputs/Final_Kmers_CC.csv')
df_kmers_BP=pd.read_csv('data/_Outputs/Final_Kmers_BP.csv')
df_kmers_MF=pd.read_csv('data/_Outputs/Final_Kmers_MF.csv')
```

```python
#df_unirep_MF['Method']='Unirep'
```

### Add a column to specifiy the GO category

```python
df_deepGo_CC['GO']='CC'
df_deepGo_MF['GO']='MF'
df_deepGo_BP['GO']='BP'

df_TripletProt_CC['GO']='CC'
df_TripletProt_MF['GO']='MF'
df_TripletProt_BP['GO']='BP'

df_unirep_CC['GO']='CC'
df_unirep_MF['GO']='MF'
df_unirep_BP['GO']='BP'

df_kmers_CC['GO']='CC'
df_kmers_MF['GO']='MF'
df_kmers_BP['GO']='BP'
```

### Aggregate allthe results

```python
df_total=pd.concat([df_deepGo_MF,df_deepGo_BP,df_deepGo_CC,df_TripletProt_MF,df_TripletProt_BP,df_TripletProt_CC,df_unirep_MF,df_unirep_BP,df_unirep_CC,df_kmers_MF,df_kmers_BP,df_kmers_CC])
```

### Exclude non discriminative tresholds  

```python
df_total=df_total[df_total.treshold<=0.3]
```

# F1 (Micro)

```python
df_F1_micro=df_total.iloc[:,[2,4,5]]
```

```python
df_F1_micro.head()
```

```python
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid", palette="muted")
from matplotlib.legend import Legend
```

```python
g=sns.swarmplot(x="GO", y="F1 (micro)", hue="Method",
              palette=['greenyellow','b','cyan','limegreen'], data=df_F1_micro,size=10)
g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
g.plot([], [], ' ', label="GO: ")
g.plot([], [], ' ', label="MF (Molecular Funstions)")
g.plot([], [], ' ', label="BP (Biological Processes)")
g.plot([], [], ' ', label="CC (Cellular Components)")
g.plot([], [], ' ', label=" ")
g.plot([], [], ' ', label="Method: ")
g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
```
